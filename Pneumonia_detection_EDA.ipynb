````python
def  extractInfo(dataFrame):
  patient_objects = dict()
  df = dataFrame.copy(True)
  for index, row in df.iterrows():
    # print(row)
    if not np.isnan(row['x']):
      pid = row['patientId']
      x1 = row['x']
      y1 = row['y']
      x2 = x1 + row['width']
      y2 = y1 + row['height']
      height = row['height']
      width = row['width']
      target = row['Target']
      lungType = row['class']
      new_box = [x1,y1,x2,y2,width,height]
      if pid in patient_objects:
        patient_obj = patient_objects[pid]
        old_box = patient_obj['boxes'][0]
        box =  [old_box,new_box]
        temp_obj = {
            'patientId': pid,
            'Target': target,
            'boxes': box,
            'Class': lungType
                    }
        patient_objects[pid] = temp_obj
      else:
        temp_obj = {
            'patientId': pid,
            'Target': target,
            'boxes': [new_box],
            'Class': lungType
                    }
        patient_objects[pid] = temp_obj
      # break
  return patient_objects
````

````python
# Extract image, target and bounding box. And prepare image path.
def extractImages(foldername,data):
    X_image_train = []
    y_image_train = np.zeros((len(data),4))
    y_train_Target = np.zeros((len(data),1))
    for index,row in data.iterrows():
        name = row[0]
        x1 = int(row[1])
        y1 = int(row[2])

        path = os.path.join( base_path,foldername,name)
        path = path+'.png'
        
        if os.path.isfile(path):
            img = cv2.imread(path)
            image_width = img.shape[1]
            image_height = img.shape[0]

            width = int(row[3])
            height = int(row[4])
            target = int(row[5])
            if width != 0 :
                y_image_train[index,0] = x1* image_size/image_width
                y_image_train[index,1] = y1* image_size/image_height
                y_image_train[index,2] = ((width+x1)-x1)* image_size/image_width
                y_image_train[index,3] = ((height+y1)-y1)* image_size/image_width
            else:
                y_image_train[index,0] = 0
                y_image_train[index,1] = 0
                y_image_train[index,2] = 0
                y_image_train[index,3] = 0

            y_train_Target[index] = target
            X_image_train.append(path)
            
    return (X_image_train,y_image_train,y_train_Target)
````

````python
#Do preprocessing image using preptrained VGG16.
def preprocessImage(data):
    processed_data = []
    for i,f in enumerate(data):
        img = Image.open(f)
        img = img.resize((224, 224)) # Resize image
        img = img.convert('RGB')
        processed_data.append(preprocess_input(np.array(img, dtype=np.float32)))
        img.close()
    return processed_data
````
